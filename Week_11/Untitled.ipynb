```python

class Trapezium():
    def __init__(self, h, b1, b2):
        self.h = h
        self.b1 = b1
        self.b2 = b2
    def calculate(self):
        solve = self.h/2 * (self.b1 + self.b2)
        print(f"Your answer is {solve}")
class Rhombus():
    d1 = input("Enter the diagonal 1: ")
    d2 = input("Enter the diagonal 2: ")
    def __init__(self, d1, d2):
        self.d1 = d1
        self.d2 = d2
    def calculate(self):
        solve = 0.5 * self.d1 * self.d2
        print(f"Your answer is {solve}")
class Parallelogram():

    def __init__(self, b, alt):
        self.b = b
        self.alt = alt
    def calculate(self):
        
        solve = self.b * self.alt
        print(f"Your answer is {solve}")
class Cube():
    
    def __init__(self, l):
        self.l = l
    def calculate(self):
        
        solve = 6 * (self.l * self.l)
        print(f"Your answer is {solve}")
class Cylinder():
    
    
    def __init__(self, h, r):
        self.h = h
        self.r = r
    def calculate(self):
        p = 3.142
        solve = self.p * (self.r * self.r) * self.h
        print(f"Your answer is {solve}")
        
        
choose = input("Enter the calculation: \nArea of Trapezium enter T\nArea of Rhombus enter R\nArea of Parallelogram enter P\nArea of Cube enter C\nVolume of Cylinder enter V")
if choose == "T":
    h = int(input("Enter the height: "))
    b1 = int(input("Enter the base 1: "))
    b2 = int(input("Enter the base 2: "))
    math = Trapezium(h, b1, b2)
    math.calculate()
elif choose == "P":
    b = int(input("Enter the base: "))
    alt = int(input("Enter the altitude: "))
    math = Parallelogram(b, alt)
    math.calculate()
elif choose == "R":
    d1 = int(input("Enter the diagonal 1: "))
    d2 = int(input("Enter the diagonal 2: "))
    math = Rhombus(d1, d2)
    math.calculate()
elif choose == "C":
    l = int(input("Enter the length of side: "))
    math = Cube(l)
    math.calculate()
elif choose == "V":
    h = int(input("Enter the height: "))
    r = int(input("Enter the radius: "))
    math = Cylinder(h, r)
    math.calculate()
else:
    print("Enter one of the quantities above")
    
```
